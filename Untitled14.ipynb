```python
import os

# Read the existing script
with open('train.py', 'r') as f:
    lines = f.readlines()

# Identify the actual folder names present in the directory
actual_classes = [d for d in os.listdir('data/plantvillage') if os.path.isdir(os.path.join('data/plantvillage', d))]

# Replace the TARGET_CLASSES list in the script content
new_lines = []
skip = False
for line in lines:
    if 'TARGET_CLASSES = [' in line:
        new_lines.append('TARGET_CLASSES = ' + str(actual_classes) + '\n')
        skip = True
    elif skip and ']' in line:
        skip = False
        continue
    if not skip:
        new_lines.append(line)

# Write the modified script back
with open('train.py', 'w') as f:
    f.writelines(new_lines)

print(f'Updated train.py to include {len(actual_classes)} classes found in data/plantvillage.')

# Now run the training again
print('--- Starting Training (5 Epochs) ---')
!python train.py --data_dir data/plantvillage --epochs 5 --batch_size 32
```

```python
import torch
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
from torchvision import datasets, transforms, models
from torch.utils.data import DataLoader
import os

def plot_confusion_matrix(model_path, data_dir):
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

    # 1. Define Transform
    transform = transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])

    # 2. Load Dataset and Model
    dataset = datasets.ImageFolder(data_dir, transform=transform)
    classes = dataset.classes

    # Use a subset for faster evaluation if dataset is huge, or full for accuracy
    indices = np.random.choice(len(dataset), min(1000, len(dataset)), replace=False)
    subset = torch.utils.data.Subset(dataset, indices)
    loader = DataLoader(subset, batch_size=32, shuffle=False)

    model = models.resnet18()
    model.fc = torch.nn.Linear(model.fc.in_features, len(classes))
    model.load_state_dict(torch.load(model_path, map_location=device))
    model.to(device)
    model.eval()

    # 3. Collect Predictions
    all_preds = []
    all_labels = []

    with torch.no_grad():
        for inputs, labels in loader:
            inputs = inputs.to(device)
            outputs = model(inputs)
            preds = torch.argmax(outputs, dim=1)
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.numpy())

    # 4. Generate and Plot Matrix
    cm = confusion_matrix(all_labels, all_preds)
    plt.figure(figsize=(12, 10))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=classes, yticklabels=classes)
    plt.title('Overall Model Confusion Matrix')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.xticks(rotation=90)
    plt.show()

plot_confusion_matrix('outputs/best_model.pth', 'data/plantvillage')
```

```python
import os
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms, models
from torch.utils.data import DataLoader

# 1. Create the train.py script from scratch since it is missing
train_script_content = """
import os
import argparse
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms, models
from torch.utils.data import DataLoader, Subset
import json

TARGET_CLASSES = [] # To be filled dynamically

class FilteredDataset(datasets.ImageFolder):
    def __init__(self, root, transform=None):
        super().__init__(root, transform=transform)
        self.samples = [s for s in self.samples if os.path.basename(os.path.dirname(s[0])) in TARGET_CLASSES]
        self.imgs = self.samples

def train(args):
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    transform = transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])

    dataset = FilteredDataset(args.data_dir, transform=transform)
    loader = DataLoader(dataset, batch_size=args.batch_size, shuffle=True)

    model = models.resnet18(weights='IMAGENET1K_V1')
    model.fc = nn.Linear(model.fc.in_features, len(TARGET_CLASSES))
    model = model.to(device)

    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=0.001)

    os.makedirs('outputs', exist_ok=True)

    for epoch in range(args.epochs):
        model.train()
        running_loss = 0.0
        for inputs, labels in loader:
            inputs, labels = inputs.to(device), labels.to(device)
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            running_loss += loss.item()
        print(f'Epoch {epoch+1}/{args.epochs}, Loss: {running_loss/len(loader)}')

    torch.save(model.state_dict(), 'outputs/best_model.pth')

if __name__ == '__main__':
    parser = argparse.ArgumentParser()
    parser.add_argument('--data_dir', type=str, required=True)
    parser.add_argument('--epochs', type=int, default=5)
    parser.add_argument('--batch_size', type=int, default=32)
    args = parser.parse_args()
    train(args)
"""

with open('train.py', 'w') as f:
    f.write(train_script_content)

# 2. Scan subdirectories to identify actual class folders
plantvillage_path = 'data/plantvillage'
actual_classes = [d for d in os.listdir(plantvillage_path) if os.path.isdir(os.path.join(plantvillage_path, d))]
print(f'Found {len(actual_classes)} classes for training.')

# 3. Read and programmatically update 'train.py'
with open('train.py', 'r') as f:
    lines = f.readlines()

new_lines = []
for line in lines:
    if 'TARGET_CLASSES = []' in line:
        new_lines.append(f'TARGET_CLASSES = {actual_classes}\n')
    else:
        new_lines.append(line)

with open('train.py', 'w') as f:
    f.writelines(new_lines)
print('Successfully updated train.py with dynamic class list.')

# 4. Execute training script
print('--- Starting Training (5 Epochs) ---')
!python train.py --data_dir data/plantvillage --epochs 5 --batch_size 32

# 5. Confirm output generation
if os.path.exists('outputs/best_model.pth'):
    print('\nSuccess: outputs/best_model.pth has been generated.')
else:
    print('\nError: outputs/best_model.pth was not found.')
```

```python
import os

# 1. Update gradcam.py to fix the endswith typo and use lowercase for robust checking
gradcam_script_content = """
import os
import argparse
import torch
import torch.nn as nn
import numpy as np
import matplotlib.pyplot as plt
from torchvision import models, transforms, datasets
from PIL import Image
import cv2

def get_gradcam(model, input_tensor, target_layer):
    features = []
    gradients = []
    def save_gradient(grad): gradients.append(grad)
    def save_features(module, input, output):
        features.append(output)
        output.register_hook(save_gradient)
    handle = target_layer.register_forward_hook(save_features)
    output = model(input_tensor)
    target_class = output.argmax(dim=1).item()
    model.zero_grad()
    output[0, target_class].backward()
    handle.remove()
    weights = torch.mean(gradients[0], dim=(2, 3), keepdim=True)
    cam = torch.sum(weights * features[0], dim=1).squeeze()
    cam = np.maximum(cam.detach().cpu().numpy(), 0)
    cam = cv2.resize(cam, (224, 224))
    cam = (cam - cam.min()) / (cam.max() - cam.min() + 1e-8)
    return cam, target_class

def main(args):
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    model = models.resnet18()
    model.fc = nn.Linear(model.fc.in_features, 15)
    model.load_state_dict(torch.load(args.model, map_location=device))
    model.to(device)
    model.eval()

    transform = transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])

    os.makedirs('outputs/heatmaps', exist_ok=True)
    valid_extensions = ('.jpg', '.jpeg', '.png', '.JPG')
    # Fixed typo: endswith
    image_files = [os.path.join(args.plantvillage_dir, f) for f in os.listdir(args.plantvillage_dir) if f.lower().endswith(valid_extensions)]

    indices = np.random.choice(len(image_files), min(args.n_samples, len(image_files)), replace=False)

    for i, idx in enumerate(indices):
        img_path = image_files[idx]
        img_pil = Image.open(img_path).convert('RGB')
        img_tensor = transform(img_pil).unsqueeze(0).to(device)

        cam, pred = get_gradcam(model, img_tensor, model.layer4[-1])

        raw_img = np.array(img_pil.resize((224, 224))) / 255.0
        heatmap = cv2.applyColorMap(np.uint8(255 * cam), cv2.COLORMAP_JET)
        heatmap = np.float32(heatmap) / 255
        cam_img = heatmap + np.float32(raw_img[:, :, ::-1])
        cam_img = cam_img / np.max(cam_img)

        plt.figure(figsize=(10, 5))
        plt.subplot(1, 2, 1); plt.imshow(raw_img); plt.title('Original Image'); plt.axis('off')
        plt.subplot(1, 2, 2); plt.imshow(cam_img[:, :, ::-1]); plt.title('Grad-CAM Analysis'); plt.axis('off')

        plt.savefig(f'outputs/heatmaps/targeted_sample_{i}.png')
        plt.close()
    print(f'Successfully generated targeted Grad-CAM visualizations.')

if __name__ == "__main__":
    parser = argparse.ArgumentParser()
    parser.add_argument('--model', type=str, required=True)
    parser.add_argument('--plantvillage_dir', type=str, required=True)
    parser.add_argument('--n_samples', type=int, default=4)
    args = parser.parse_args()
    main(args)
"""

with open('gradcam.py', 'w') as f:
    f.write(gradcam_script_content)
```

## Robustness Check with Augmentations

### Subtask:
Execute 'augmentation_analysis.py' to evaluate how the model handles image variations across 20 samples.


**Reasoning**:
I will create and execute the 'augmentation_analysis.py' script to evaluate the model's robustness by applying various image transformations to random samples from the dataset and checking for classification consistency.



```python
import os

# 1. Create the augmentation_analysis.py script
augmentation_script_content = """
import os
import argparse
import torch
import torch.nn as nn
import numpy as np
from torchvision import models, transforms, datasets
from PIL import Image
import json

def main(args):
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

    # Dynamically determine number of classes
    classes = sorted([d for d in os.listdir(args.plantvillage_dir) if os.path.isdir(os.path.join(args.plantvillage_dir, d))])
    num_classes = len(classes)

    # Load Model
    model = models.resnet18()
    model.fc = nn.Linear(model.fc.in_features, num_classes)
    model.load_state_dict(torch.load(args.model, map_location=device))
    model.to(device)
    model.eval()

    # Define Augmentations
    augmentations = {
        'Original': transforms.Compose([transforms.Resize((224, 224)), transforms.ToTensor()]),
        'Horizontal Flip': transforms.Compose([transforms.Resize((224, 224)), transforms.RandomHorizontalFlip(p=1.0), transforms.ToTensor()]),
        'Rotation': transforms.Compose([transforms.Resize((224, 224)), transforms.RandomRotation(30), transforms.ToTensor()]),
        'Color Jitter': transforms.Compose([transforms.Resize((224, 224)), transforms.ColorJitter(brightness=0.5, contrast=0.5), transforms.ToTensor()])
    }

    norm = transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])

    # Select random samples
    full_dataset = datasets.ImageFolder(args.plantvillage_dir)
    indices = np.random.choice(len(full_dataset), args.n_samples, replace=False)

    results = []

    for idx in indices:
        path, label = full_dataset.samples[idx]
        img = Image.open(path).convert('RGB')
        sample_results = {'image_path': path, 'actual_class': classes[label], 'predictions': {}}

        for aug_name, transform in augmentations.items():
            img_tensor = transform(img).unsqueeze(0).to(device)
            normalized_tensor = norm(img_tensor)

            with torch.no_grad():
                output = model(normalized_tensor)
                pred_idx = output.argmax(dim=1).item()
                sample_results['predictions'][aug_name] = classes[pred_idx]

        results.append(sample_results)

    # Save Report
    output_path = 'outputs/augmentation_report.json'
    with open(output_path, 'w') as f:
        json.dump(results, f, indent=4)

    print(f'Successfully analyzed {args.n_samples} samples. Report saved to {output_path}.')

if __name__ == "__main__":
    parser = argparse.ArgumentParser()
    parser.add_argument('--model', type=str, required=True)
    parser.add_argument('--plantvillage_dir', type=str, required=True)
    parser.add_argument('--n_samples', type=int, default=20)
    args = parser.parse_args()
    main(args)
"""

with open('augmentation_analysis.py', 'w') as f:
    f.write(augmentation_script_content)

# 2. Execute the script
print('--- Running Augmentation Analysis ---')
!python augmentation_analysis.py --model outputs/best_model.pth \
                                 --plantvillage_dir data/plantvillage \
                                 --n_samples 20

# 3. Verify output
if os.path.exists('outputs/augmentation_report.json'):
    print('\nSuccess: outputs/augmentation_report.json has been generated.')
```
